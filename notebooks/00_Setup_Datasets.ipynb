```python
from google.colab import drive
import os

drive.mount('/content/drive')


BASE_DIR = '/content/drive/MyDrive/PRISM'
DATASETS = {
    'pcam':          f'{BASE_DIR}/datasets/pcam',
    'crc':           f'{BASE_DIR}/datasets/crc',
    'mhist':         f'{BASE_DIR}/datasets/mhist',
    'bracs':         f'{BASE_DIR}/datasets/bracs',
    'panda':         f'{BASE_DIR}/datasets/panda',
    'spider_breast': f'{BASE_DIR}/datasets/spider_breast',
}

for name, path in DATASETS.items():
    os.makedirs(path, exist_ok=True)
    print(f"Created: {path}")
```

```python
from torchvision.datasets import PCAM

print("Downloading PCam to Drive...")
for split in ['train', 'val', 'test']:
    PCAM(root=DATASETS['pcam'], split=split, download=True)
    print(f"  {split} done")
print("PCam complete!")
```

```python
import urllib.request
import zipfile

CRC_URL = "https://zenodo.org/record/1214456/files/NCT-CRC-HE-100K.zip"
CRC_ZIP = f"{DATASETS['crc']}/NCT-CRC-HE-100K.zip"

print("Downloading CRC (Kather texture tiles)...")
urllib.request.urlretrieve(CRC_URL, CRC_ZIP)
print("Extracting...")
with zipfile.ZipFile(CRC_ZIP, 'r') as z:
    z.extractall(DATASETS['crc'])
os.remove(CRC_ZIP)
print("CRC complete!")
```

```python
import requests
import os
import zipfile

LUNGHIST_DIR = '/content/drive/MyDrive/PRISM/datasets/lunghist700'
os.makedirs(LUNGHIST_DIR, exist_ok=True)

# Figshare API ile doğru indirme
FIGSHARE_ID = "25459174"
API_URL = f"https://api.figshare.com/v2/articles/{FIGSHARE_ID}/files"


response = requests.get(API_URL)
files = response.json()

for f in files:
    print(f"Dosya: {f['name']} - {f['size']/1e6:.1f} MB")
    print(f"Link: {f['download_url']}")

# Her dosyayı indir
for f in files:
    out_path = f"{LUNGHIST_DIR}/{f['name']}"
    if not os.path.exists(out_path):
        print(f"\nDownloading: {f['name']}...")
        r = requests.get(f['download_url'], stream=True)
        with open(out_path, 'wb') as fp:
            for chunk in r.iter_content(chunk_size=8192):
                fp.write(chunk)
        print(f"Tamam: {f['name']}")

print("\nTüm dosyalar indirildi!")
os.listdir(LUNGHIST_DIR)
```
